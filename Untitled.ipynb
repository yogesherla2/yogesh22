```python
# Define a custom color map for decision regions
from matplotlib.colors import ListedColormap

# Custom colormap for plotting
custom_cmap = ListedColormap(['lightblue', 'lightgreen', 'lightpink', 'wheat', 'lavender'])

def plot_decision_boundary_alternative(X, y, clf, title='Decision Boundary'):
    """
    Visualizes the decision boundary of a classifier in a custom way.

    Parameters:
    - X: Feature matrix (numpy array)
    - y: Labels (numpy array)
    - clf: Trained classifier
    - title: Title for the plot
    """
    # Create a grid of points covering the feature space
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                         np.linspace(y_min, y_max, 500))

    # Predict classes for each point on the grid
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Create the figure
    plt.figure(figsize=(8, 6))

    # Plot decision regions
    plt.contourf(xx, yy, Z, alpha=0.6, cmap=custom_cmap)

    # Scatter plot of data points
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm, s=30)
    legend1 = plt.legend(*scatter.legend_elements(), title="Classes")
    plt.gca().add_artist(legend1)

    # Adding titles and labels
    plt.title(title, fontsize=14)
    plt.xlabel('Feature 1', fontsize=12)
    plt.ylabel('Feature 2', fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()
```

```python
def exploreSVMKernel(dataset):
    # Assuming the dataset is extracted and you provide its name (e.g., 'Aggregation')
    file_path = os.path.join(extract_folder, f"{dataset}.txt")  # Build the file path from the extracted folder

    # Read the dataset with the appropriate delimiter
    try:
        data = pd.read_csv(file_path, sep="\t", skiprows=7, header=None, names=['X1', 'X2', 'y'])
        print(f"Dataset loaded for {dataset}.")
        print(data.head())  # Show the first few rows for inspection
    except Exception as e:
        print(f"Error loading dataset {dataset}: {e}")

        print(f"Calculating the accuracies and plotting the decision boundaries of the Linear, Polynomial, RBF and Sigmoid kernels for the >{dataset}< dataset.")

    # Creating test and train sets to determine the accuracy of the different kernels
    X = data.drop(columns='y').values
    y = data['y'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2984)

    for kernel in ['linear', 'poly', 'rbf', 'sigmoid']:
        Model = OneVsRestClassifier(SVC(kernel=kernel))
        Model.fit(X_train, y_train)

        Predictions = Model.predict(X_test)
        Accuracy = np.mean(Predictions == y_test)*100
        print(f"\n The accuracy of the SVM >{kernel}< model on the >{dataset}< dataset is: {Accuracy:.4f}%")
        plot_decision_boundary(X, y, Model)
```

```python
def plot_results_kernels_with_gaps(results):
    """Plot accuracies with separate bars for kernels and gaps between them."""
    datasets = list(results.keys())  # Dataset names
    kernels = ['linear', 'poly', 'rbf', 'sigmoid']  # Kernels
    n_datasets = len(datasets)

    # Adjust spacing for bar positions
    bar_positions = []
    bar_width = 0.15  # Width of each bar
    group_gap = 0.6  # Gap between kernel groups

    for i, kernel in enumerate(kernels):
        bar_positions.append(i * (n_datasets * bar_width + group_gap))  # Position for each kernel group

    plt.figure(figsize=(14, 8))

    # Loop through datasets and plot bars for each kernel
    for dataset_index, dataset in enumerate(datasets):
        accuracies = [results[dataset][kernel] for kernel in kernels]
        positions = [pos + dataset_index * bar_width for pos in bar_positions]  # Shift bars within each kernel group
        plt.bar(positions, accuracies, bar_width, label=dataset)

    # Formatting the plot
    plt.xlabel('Kernels', fontsize=14)
    plt.ylabel('Accuracy (%)', fontsize=14)
    plt.title('Accuracy of SVM Kernels Across Datasets', fontsize=16)
    plt.xticks([pos + (n_datasets - 1) * bar_width / 2 for pos in bar_positions], kernels, fontsize=12)  # Center ticks
    plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines along y-axis
    plt.legend(title="Datasets", fontsize=12)
    plt.tight_layout()
    plt.show()

# Call the updated plotting function
plot_results_kernels_with_gaps(results)
```
